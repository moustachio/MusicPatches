```python
data = pd.read_csv('total_user_total_threshold_data_tmp4to6.csv', header = 0)
print data.head()
```

```python
#load the second file where the thresholds are at 0.7 yo 0.9
data1 = pd.read_csv('total_user_total_threshold_data_tmp7to9.csv', header = 0)
print data1.head()
```

```python
#print data.loc[,'user_id'].unique_counts
unique_users = data.loc[:,'user_id'].unique()
```

```python
threshold_set = [0.4,0.5,0.6]
threshold_set1 = [0.7,0.8,0.9]
```

```python
# define a new variable : scrobble length
scro_len_set = []
for each_user in unique_users:
    #print data4[data4.user_id == each_user].shape[0]
    scro_len_set.append(data4[data4.user_id == each_user].shape[0])
```

```python
#assign the id, scro_len, threshold and etc to one dataframe
patch_df = pd.DataFrame(columns = ['user_id','scro_len'])
patch_df['user_id'] = unique_users
patch_df['scro_len'] = scro_len_set
#print patch_df
patch_df = patch_df.set_index('user_id')
```

```python
print patch_df
print patch_df.index
```

```python
# compute the number of patches of each subject at each threshold  at threshold_set 
#patch_df
for thres in threshold_set:
    print thres
    # there is a bug in pandas here!! when thres = 0.6, it doesnot work for the dataframe to extract data
    if thres == 0.6:
        data_thres = data[data.threshold >0.6]
    else:    
        data_thres = data[data.threshold == thres]
    data_thres_agg = data_thres.groupby('user_id').agg(max)
    columnName = 'numPatch' + str(thres)
    data_thres_agg_max = pd.DataFrame(columns = [columnName], index = patch_df.index)
    #for each_user in unique_users[:20]:--for test
    for each_user in unique_users:
        data_thres_agg_max.loc[each_user] = data_thres_agg.loc[each_user].cluster_seq
    patch_df[columnName] = data_thres_agg_max[columnName]
print patch_df
```

```python
# compute the number of patches of each subject at each threshold  at threshold_set1 
#patch_df
for thres in threshold_set1:
    print thres
    # there is a bug in pandas here!! when thres = 0.7, it doesnot work for the dataframe to extract data
    if thres == 0.7:
        data_thres = data1[data1.threshold < 0.75]
    else:    
        data_thres = data1[data1.threshold == thres]
    data_thres_agg = data_thres.groupby('user_id').agg(max)
    columnName = 'numPatch' + str(thres)
    data_thres_agg_max = pd.DataFrame(columns = [columnName], index = patch_df.index)
    #for each_user in unique_users[:20]:
    for each_user in unique_users:
        data_thres_agg_max.loc[each_user] = data_thres_agg.loc[each_user].cluster_seq
    patch_df[columnName] = data_thres_agg_max[columnName]

print patch_df
```

```python
#compute the #S/#P for each thres
threshold_total_set = threshold_set + threshold_set1
print threshold_total_set
for thres in threshold_total_set:
    columnName = 'numPatch' + str(thres)
    newColumnName = columnName + '_Scro_PerPatch'
    patch_df[newColumnName] = patch_df['scro_len']/patch_df[columnName]
print patch_df
```

```python
#compute #p/(#si/#s_min) for each thres
#first get the min value of # scrobbles
min_num_scro_len = patch_df.scro_len.min()

for thres in threshold_total_set:
    columnName = 'numPatch' + str(thres)
    newColumnName = 'norm_numPatch' + str(thres)
    patch_df[newColumnName] = patch_df[columnName]/(patch_df['scro_len']/min_num_scro_len)
print patch_df
```

```python
#NOTE THAT: pandas series will allign the assignment according to the index of both dataframe and series, in this way you will get all Nans
patch_stat['norm_numPatch_aver'] = patch_df.apply(mean).iloc[-6:].values
patch_stat['norm_numPatch_std'] = patch_df.apply(std).iloc[-6:].values
```

```python
#errors in the plot has to be: str of column names of plotted df; list,tuple or nparray; or dataframe or dict of errors--cannot be series!
errors = np.asarray(patch_stat.norm_numPatch_std/sqrt(patch_df.shape[0]))
ax =patch_stat.plot(x = 'threshold', y = 'norm_numPatch_aver', yerr= errors, legend=False, title = 'Changes of Normalized Number of Patches by Threshold')
ax.set_xlim([0.3,1])
ax.set_ylim([15,40])
ax.set_xlabel('Threshold')
ax.set_ylabel('Normalized Number of Patches')
```

```python
errors1 = np.asarray(patch_stat.scro_perPatch_std/sqrt(patch_df.shape[0]))
ax =patch_stat.plot(x = 'threshold', y = 'scro_perPatch_aver', yerr= errors1, legend = False, title =  'Changes of Number of Scrobbles per Patch by Threshold')
ax.set_xlim([0.3,1])
ax.set_ylim([0,22])
ax.set_xlabel('Threshold')
ax.set_ylabel('Number of Scrobbles per Patch')
```
