```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
%pylab inline
import sklearn as sk
```

```python
data = []
id_lst = []
for line in open('simSeqs.tsv').readlines():
#     i +=1
#     print len(line)  
#     if i >3 :
#         break
#     print i
    
    each_row = line.split()
    each_id = each_row[0]
    each_row = each_row[1:]
    id_lst.append(each_id)
    each_row = [float(element) for element in each_row if element != 'None']
    #print each_row
    data.append(each_row)
```

```python
#numpy could handle different length of rows from a list, and transfer it from list to array, but cannot do that when reading automaticaly
#data = np.genfromtxt('simSeqs.tsv', dtype = 'float32',missing_values = 'None')
data = np.array(data)
print data.shape
print data.size
id_lst = np.array(id_lst)
# save the data of  simlarity sequence and also listener's id in one npz file
#----------------------------------------------
#np.savez('simSeqs', data = data, id = id_lst)
```

```python
#randomly plot several subjects to show the individual difference in terms of the similarity distribution across all the listeners
num_users = data.shape[0]
#randint: exclusive for high, random_integers: inclusive both ends
listener_set = np.random.randint(0, num_users, 20)
print listener_set
```

```python
num_figure = len(listener_set)
num_col = 2
num_row = num_figure/num_col
figure1, axes = plt.subplots(num_row, num_col, figsize=(16,40), dpi= 80)

i = 0
for row in range(num_row):
    for col in range(num_col):
        axes[row, col].hist(data[listener_set[i]])
        axes[row, col].set_title('Listener ' + str(listener_set[i]))
        i +=1
```
